```python
import time
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
Num = input("è¼¸å…¥ä½ çš„å°æ•¸ :")
```

```python
MIN = 8 
dic={}
log10_str = [ "10^{}".format(i) for i in range(2,int(Num)+1)]

def calPI(s):
    ndata = int(10 ** MIN) if s>MIN else int(10 ** s)
    forLoop = int(10 ** (s-MIN)) if s>MIN else 1
    total = 0.0 
    for _ in range(forLoop):
        x = np.random.uniform(-1,1,ndata)
        y = np.random.uniform(-1,1,ndata)
        z = np.sqrt(x ** 2 + y ** 2)
        total += (z<=1).sum()
    pi=4*(total/ndata) if s<=MIN else 4*(total/(10 ** s)) 
    return pi

def main():
    np.random.seed(int(time.time()))
    
    for i in range(2,int(Num)+1):
        dic[i]=round(calPI(i),7)

if __name__ == '__main__':
    main()
```

```python
print("PI dictionary (2 - {}):".format(Num))
print(dic)
```
