```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg


df = pd.read_csv('0050.csv')
df['Ups_Downs']=0
for i in (range(df.shape[0])):
    if i==0:
        continue
    a=df.at[i,'Close']
    b=df.at[i-1,'Close']
    df.at[i,'Ups_Downs']=(a-b)/a*100
print(df[2400::])
data = df['Ups_Downs'].to_numpy(dtype='f4')

model = AutoReg(data, 300, old_names=False)
res = model.fit()
#print(res.summary())
#fig = plt.figure(figsize=(20,12))
fig = res.plot_predict(2400, 2500)
```
